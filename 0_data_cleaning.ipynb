Imports

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pathlib import Path

dataset_name = 'risk_factors_cervical_cancer.csv'
dataset_path = Path('./datasets') / dataset_name
```

## **Loading of data set and exploring the data**


```python
data = pd.read_csv(dataset_path, header=0)
data
```

## **Data cleaning**

```python
data.isnull().any()
```

```python
data = data.replace('?', np.nan)
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'], axis=1)
```

```python
data.shape
```

**Deleting null values**

```python
data = data.dropna(axis=0)
```

```python
data.shape
```

```python
data = data._convert(numeric=True)
```
